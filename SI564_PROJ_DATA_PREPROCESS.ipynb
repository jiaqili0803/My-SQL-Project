```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('original_data.csv')
```

```python
df
```

```python
np.max(df['IMDB Score'])
```

```python
np.min(df['IMDB Score'])
```

```python
plt.hist(df["IMDB Score"])

# Add chart title and axis labels
plt.title("IMDB Score Distribution")
plt.xlabel("Score")
plt.ylabel("Count")

# Display the chart
plt.show()
```

```python
unique_genres = sorted(df['Genre'].unique())
genre_to_id = {genre: idx for idx, genre in enumerate(unique_genres)}
df['Genre_id'] = df['Genre'].map(genre_to_id)
```

```python
df.to_csv('updated_genre_talbe.csv', index=False)
```

```python
df
```

```python
print('there are in total: ', max(df['Genre_id']), 'genres')
```
